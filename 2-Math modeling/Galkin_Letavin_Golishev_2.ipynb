```python
import matplotlib.pyplot as plt
import numpy as np
import math as m
from sympy import *
from sympy.abc import z

count = 0
N = 1
Sn = 1/17
S = 0
E = 0.001

my_list = list()
my_list2 = list()

an = 1 / (z**2 + 8 * z + 8)
my_list.append(Sn)
my_list2.append(Sn)

while((Sn - S) > E):
    N += 1
    Sn += an.subs(z, N)
    my_list.append(Sn)
    my_list2.append(an.subs(z, N))
    if N >= 2:
        S = my_list[N-2]
        
print (Sn)
print (S)
print(N)

x = list()
for i in my_list:
    count += 1
    x.append(count)
plt.figure(1)
plt.plot(x, my_list)
plt.figure(2)
plt.plot(x, my_list2)
plt.show()
    
```
