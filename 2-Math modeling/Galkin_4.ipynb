```python
import matplotlib.pyplot as plt
import numpy as np
import math as m
from sympy import *

def matrix(): #Функция задания матрицы 
    coef_a = 10.
    A = np.zeros((100, 100))
    F = np.zeros(100)
    for i in range(0,100,1):
        F[i] = 100 - i
        for j in range(0,100,1):
            if(i == 0):
                A[i][j] = 1
            elif((i > 0) & (i <= 98)):
                if(i == j):
                    A[i][j] = coef_a * 1.
                    A[i][j-1] = 1.
                    A[i][j+1] = 1.
                    continue
            else:
                if(j > 97):
                    A[i][j] = 1.
    return A, F
A,F = matrix() #Задание матрицы

print("Матрица коэффициентов - ",'\n',A)
print("Значения - ",F)

X = np.zeros(100)
saveA = A.copy() # Копирование матрицы А
saveF = F.copy() # Копирование матрицы F

m = np.linalg.norm(A, 1) * np.linalg.norm(np.linalg.inv(A) , 1) # вычесление числа обусловленности      

lambdmin = 1000000000000 # Перемнная для лямбда минимальное
lambdmax = -1000000000000 # Переменная для лямбда максимальное

lambds = list() # задание пустого списка всех лямбд
lambds = np.linalg.eig(A) # задания списка всех лямбд

for k in range (1, 100, 1): #Метод Гаусса(делем верхнетреугольную матрицу А, и заполняем матрицу F новыми значениями)
    for i in range(k, 100, 1):
        aij = A[i][k - 1]
        for j in range(0, len(A[i] - 1), 1):
            A[i][j] = A[i][j] - A[k - 1][j] * aij  / A[k - 1][k - 1]
        F[i] = F[i] - F[i - 1] / A[k - 1][k - 1] * aij

for i in range(99, -1, -1): # Подставляем Х в нашу систему снизу вверх
    sum = 0.
    for j in range(100):
        sum += A[i][j] * X[j]
    X[i] = (F[i] - sum) / A[i][i]
    
R = abs(saveF - np.dot(saveA, X)) # Нахождение вектора невязки
    
for i in (lambds[0]): # нахождение минимальной и максимальной лямбды
    if (i >= lambdmax):
        lambdmax = i
    if(i <= lambdmin):
        lambdmin = i
                            
print("Лямбда минимальная  = ", lambdmin,"\n", "Лямбда максиммальная = ", lambdmax)
print ("Число обусловленности = ", m)
print("Список значений аргументов = ","\n", X) 
print("Вектор невязки = ",'\n', np.linalg.norm(R,1))
```
