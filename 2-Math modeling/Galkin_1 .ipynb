```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
from sympy import *
from sympy.abc import t

differ = diff(diff(atan(t), t))

const = 0.0007 

def pervproizv(x):
    return (m.atan(x + const) - m.atan(x - const)) / (2 * const)

def vtorayaproiz(x):
    znacheniya = list()
    for i in x: 
        znacheniya.append((pervproizv(i + const) - pervproizv(i - const)) / (2 * const))
    return znacheniya

def raznost(x, x1):
        znacheniya = list()
        count  = 0
        for i in x:
            count += 1
        for i in range(count):
            znacheniya.append(x1[i] - x[i])
        return znacheniya

x = np.arange(-10, 10, const) 
y = vtorayaproiz(x)
y2 = list()
for const in x:
    y2.append(differ.subs(t, const))
y3 = raznost(y, y2)

plt.figure(1)
plt.plot(x, y)
plt.plot(x, y2)

plt.figure(2)
plt.plot(x , y3)

plt.show()
```
